```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  # Para graficar
import seaborn as sns
import matplotlib.dates as mdates

from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('/content/drive/MyDrive/Masters in Data Sciences and Analysis/Semetre 1/Proyecto integrador/Datasets/Raw/Sentimientos_descripcion.csv')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.info()
```

```python
fig, ax = plt.subplots()

ax.bar(df['Date'], df['TB Negatives'], width=1 , label='TB Negative')
ax.bar(df['Date'], df['TextBlob Positives'], width=1, bottom=df['TB Negatives'], label='TB Positives')


ax.set_ylabel('# Tweets', fontsize=16)
ax.set_title('Distribucion sentimientos Textblob', fontsize=16)
ax.legend()
ax.xaxis.set_major_locator(mdates.AutoDateLocator())

plt.show()
```

```python
fig, ax = plt.subplots()

ax.bar(df['Date'], df['Vader Negatives'], width=1 , label='Vader Negatives')
ax.bar(df['Date'], df['Vader Positives'], width=1, bottom=df['Vader Negatives'], label='Vader Positives')


ax.set_ylabel('# Tweets', fontsize=16)
ax.set_title('Distribucion sentimientos Vader', fontsize=16)
ax.legend()
ax.xaxis.set_major_locator(mdates.AutoDateLocator())

plt.show()
```
