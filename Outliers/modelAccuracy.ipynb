```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
trainingDays = [1, 3, 5, 10, 15]

datosDic = {
    '1d':[],
    '3d':[],
    '5d':[],
    '10d':[],
    '15d':[]
    }

datos = pd.DataFrame(datosDic)

for i in trainingDays:
    datosDic[str(i) + 'd'] = pd.read_pickle('data/conf_list_training_'+ str(i) + 'd' + '.pkl')


dFro = []
dFroDic = pd.DataFrame({
    '1d':[],
    '3d':[],
    '5d':[],
    '10d':[],
    '15d':[]
    })

for i in trainingDays:
    dFro = []
    for j in datosDic[str(i) + 'd'].values:
        fro = np.linalg.norm(j[0][0] - np.eye(2), 'fro')
        dFro.append(fro)
    dFroDic[str(i) + 'd'] = dFro
    print('Training Days: ', str(i) + 'd', ' || Mínima distancia: ', min(dFroDic[str(i) + 'd']), ' || Índice modelo: ', list(dFroDic[str(i) + 'd']).index(min(dFroDic[str(i) + 'd'])))

plt.plot(dFroDic)
plt.xlabel('Índice del modelo'); plt.ylabel('Métrica Frobenius')
plt.minorticks_on();plt.grid(which='major', alpha=0.8); plt.grid(which='minor', alpha = 0.2)
plt.legend(['1d', '3d', '5d', '10d', '15d'])
# np.linalg.norm(test.values[0][0] - np.eye(2), 'fro')


# test.to_csv('data/prueba.csv')

```
